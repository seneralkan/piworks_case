```python
# Display the total amount earned by each employee's name and surname

amount = pd.read_sql("""
SELECT e."FirstName", SUM(p."Value") as amount
FROM payment as p
LEFT JOIN employee e on p."EmployeeID" = e."EmployeeID"
GROUP BY e."FirstName"
ORDER BY amount DESC;

""", conn)

amount
```

```python
# Display all employees that have their names starting with the J letter.

j_employee = pd.read_sql(""" 
SELECT *
FROM employee
WHERE "FirstName"
LIKE 'J%'
""", conn)
j_employee
```

### Question 2

Please write the code for identfying the 3 districts having highest number of buildings with number of buildings from shared KML files.

Please paste the link to the answer including the result and the code block using paste.tc or similar site.

KML files: https://drive.google.com/drive/folders/181sgH3elT4dcACyfisStKPVvRlROjmzs

```python
import geopandas as gpd
import fiona

gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'
df_building = gpd.read_file('./buildings.kml', driver = 'KML')
df_district = gpd.read_file('./districts.kml', driver= "KML")
df_district.columns
```

```python
sjoin = gpd.sjoin(df_building, df_district, how='inner', op='within')
sjoin.reset_index()
sjoin.drop([ 'Name_left', 'Description_left', 'index_right', 'Description_right'], axis=1, inplace=True)
```

```python
sjoin["District"] = sjoin["Name_right"]
sjoin.drop(['Name_right'], axis=1)
```

```python
count_buildings = sjoin.groupby(["District"], as_index=False)["geometry"].count()
count_buildings.sort_values("geometry", ascending=False)
```
